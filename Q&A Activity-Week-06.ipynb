```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
trade_data=pd.read_csv('Dataset_w6.xlsx - Data.csv')
```

```python
trade_data
```

## 1. Find out how many countries are there in the dataset.


### Count the no: of  countries

```python
# counting the number of countries using nunique(),so NaN values are not included in the count.
trade_data.country.nunique()
```

```python
#setting the max_row display option
pd.options.display.max_rows = 1000
```

```python
#Names of countries and GDP per capita (current US$) of each country
trade_data[['country','GDP per capita (current US$)']]
```

## 2. Examine the data by using visualizations.


```python
#Distribution of GDP per capita
trade_data['GDP per capita (current US$)'].plot(kind='kde', figsize=(10,6), title='Distribution of GDP per capita (current US$)')
```

```python
#Bar plot of Top 15countries based on International trade
tdata=trade_data.sort_values('International trade: Imports (million US$)',ascending=False).head(15)
tdata.plot.barh(x='country',y='International trade: Imports (million US$)')
```

```python
#Countplot of Region
plt.figure(figsize=(17,6))
sns.countplot(tdata['Region'], palette = 'Blues')
plt.title('Comparison of Region', fontweight = 30, fontsize = 20)
plt.xlabel('Region')
plt.ylabel('count')
plt.show()
```

```python
#Barplot of Top 15countries based on popoulation
pop_data = trade_data.sort_values('Population in thousands (2017)',ascending=False).head(15)
pop_data.plot.barh(x='country',y='Population in thousands (2017)', figsize=(16,17))
```

```python
#Barplot of Top 15 countries based on surface area
area_data = trade_data.sort_values('Surface area (km2)',ascending=False).head(15)
plt.figure(figsize=(17,6))
plt.xticks(rotation=90)
sns.barplot(x='country', y='Surface area (km2)',data=area_data)
plt.title('Surface area')
```

### 3. Find out how many countries with respect to the region are available in the dataset


```python
#countries with respect to the region 
trade_data.groupby(['Region', 'country']).size()
```

```python
#No:of countries for each region
trade_data.groupby('Region')['country'].count().reset_index()
```

### 4. On average which region has the highest 
a. GDP per capita (current US$)
b. International trade: Imports (million US$)

```python
#Based on GDP
trade_data.groupby("Region")["GDP per capita (current US$)"].mean().sort_values(ascending=False)
```

```python
#Based on International trade
trade_data.groupby("Region")["International trade: Imports (million US$)"].mean().sort_values(ascending=False)
```

### 5. Find out which region has more consistent with respect to GDP percapita (current US$)

```python
#Region has more consistent with GDP 
data=trade_data[['GDP per capita (current US$)','Region']].groupby("Region").std().sort_values(by=['GDP per capita (current US$)'],inplace=False)
data
```
